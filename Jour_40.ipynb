Un programme qui permet de calculer la fréquence des gènes par rapport aux individus, le but ici est de créer des messages personallisés en cas d'erreur

```python
def calculer_frequence(genes, individus):
    try:
        resultat = genes / individus
        print(f"Fréquence calculée : {resultat:.4f}")
    
    except ZeroDivisionError:
        print("Erreur : Impossible de calculer sur 0 individu (donnée absente).")
    
    except TypeError:
        print("Erreur : Tu as entré du texte au lieu d'un chiffre !")

    finally:
        print("Analyse de l'échantillon terminée.\n")
```

Affichage (on test les 3 types de cas)

```python
calculer_frequence(150, 10) # Cas normal
calculer_frequence(150, 0)  # Cas erreur de division
calculer_frequence(150, "Inconnu") # Cas erreur de type
```
