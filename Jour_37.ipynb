Utilisation de * devant un paramètre, pour pouvoir en utiliser un nombre indeterminé

```python
def lister_ancetres(*peuples):
    print(f"Début de l'analyse")
    for peuple in peuples:
        print(f"Analyse pour : {peuple}")
    print(f"Total de populations comparées : {len(peuples)}")
```

Affichage (grace au signe *, on peut utiliser comme ici 2 ou plusieurs arguments, sans à avoir à préciser le nombre)

```python
lister_ancetres("Vikings", "Celtes")

lister_ancetres("Yamnaya", "Chasseurs-Cueilleurs", "Fermiers Anatoliens", "Berbères")
```
