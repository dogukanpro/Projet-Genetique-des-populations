Programme pour compter le nombre d'ancêtres à chaque génération 

```python
def compter_ancetres(generation):
 
    if generation == 0:
        return 1 
    
   
    return 2 * compter_ancetres(generation - 1)
```

Exemple avec 5 générations

```python
resultat = compter_ancetres(5)
print(f"À la 5ème génération, tu as théoriquement {resultat} ancêtres directs.")
```
