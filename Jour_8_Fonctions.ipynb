La fonction qui va permettre de faire la transcription

```python
def transcription (adn):
    adn_maj = adn.upper()
    arn = adn_maj.replace("T", "U")
    return arn
```

Affichage et utilisation de la transcription

```python
sequence = "atgcgt"
resultat = transcription(sequence)

print(f"Ma séquence d'ADN : {sequence}")
print(f"La transcription en ARN : {resultat}")
```
