Utilisation de ** pour mettre dans une bibliothèque les variables

```python
def enregistrer_population(nom, **details):
    print(f"{nom}")
    
    for cle, valeur in details.items():
        print(f"{cle.capitalize()} : {valeur}")
```

Affichage

```python
enregistrer_population("Vikings", Region="Scandinavie", Ere="Médiévale\n")
enregistrer_population("Basques", Region="Pyrénées", Particularité="Isolat génétique", ADN="Mitochondrial")
```
