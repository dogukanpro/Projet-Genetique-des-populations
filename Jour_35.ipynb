Création d'une méthode qui va permettre de tranformer le score d'un peuple en pourcentage et le renvoyer, avec un message

```python
def analyser_ancetre(peuple, score_decimal):
    pourcentage = score_decimal * 100
    
    message = f"L'influence du peuple {peuple} est confirmée."
    
    return pourcentage, message

p_cent, texte = analyser_ancetre("Yamnaya", 0.23)

print(f"Résultat : {p_cent}%")
print(f"Statut : {texte}")
```
