Initialisation

```python
donnees_genetiques = ["0.85", "0.72", "MISSING", "0.91", "ERROR", "0.88"]
frequences_propres = []
```

La boucle for

```python
for donnee in donnees_genetiques:
    try:
        valeur = float(donnee)
        frequences_propres.append(valeur)
    except ValueError:
        continue
```

Calcul de la moyenne (sum pour faire la somme de ce qu'il y a, à l'intérieur de la parenthèse)

```python
moyenne_pop = sum(frequences_propres) / len(frequences_propres)
```

Affichage + utilisation de .2f pour afficher que deux chiffres après la virgule

```python
print(f"Analyseterminée.")
print(f"Échantillons valides : {len(frequences_propres)}")
print(f"Fréquence moyenne de l'allèle H-G : {moyenne_pop:.2f}")
```
