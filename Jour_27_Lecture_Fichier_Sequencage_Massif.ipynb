Utilisation de open "w", pour créer un fichier sequence adn et d'y écrire à l'intérieur

```python
with open("sequences_adn.txt", "w") as f:
    f.write("ATGC\nATTG\nGCA\n")
```

Utilisation de open "r" pour lire le fichier, puis enlever les \n (retout à la ligne) lors de la lecture, puis stockage de la lettre A avec count

```python
with open("sequences_adn.txt", "r") as fichier:
    for ligne in fichier:
        sequence = ligne.strip()
        compteur_A += sequence.count("A")
```

Affichage

```python
print(f"Nombre total d'Adénines trouvées : {compteur_A}")
```
