Initialisation

```python
hunter_gatherer = "ATGCCGTA"
early_farmer = "ATGCCCTA"
```

```python
mutations = 0
longueur = len(hunter_gatherer)
```

Le coeur du programme

```python
for i in range(longueur):
    if hunter_gatherer[i] != early_farmer[i]:
        mutations += 1
        print(f"Mutation trouvée à l'index {i} : HG({hunter_gatherer[i]}) != EF({early_farmer[i]})")
```

Calcul (de la similarité en pourcentage)

```python
similarite = ((longueur - mutations) / longueur) * 100
```

Affichage

```python
print("-" * 30)
print(f"Résultat : {mutations} mutation(s) détectée(s).")
print(f"Le taux de similarité génétique est de : {similarite}%")
```
