Création d'une méthode poour : regarder si dans une population la frequence allelique est rare ou commune

```python
def analyser_frequence(nom_population, frequence):
    if frequence < 0.05:
        statut = "RARE"
    else:
        statut = "COMMUNE"
    
    return f"Dans la population {nom_population}, cette variante est {statut}."
```

Test sur deux populations + affichage

```python
resultat_1 = analyser_frequence("Basques", 0.12)
resultat_2 = analyser_frequence("Sami", 0.02)
print(resultat_1)
print(resultat_2)
```
