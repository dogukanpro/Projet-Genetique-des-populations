Initialisation

```python
sequence1 = "ATGAA"
sequence2 = "CTGAG"
```

Je crée la fonction comparaison qui va permettre de comparer deux séquences pour voir si elles sont identique (de la même espece) ou différentes (spéciation)

```python
def comparaison(sequence1, sequence2):
    if sequence1 == sequence2 :
        return "Identique"

    else :
        return "Différents"
```

Affichage + utilisation de la fonction

```python
resultat = comparaison(sequence1, sequence2)
print(f"Résultat de la comparaison : {resultat}")
```
