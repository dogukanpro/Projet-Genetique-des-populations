Création de la bibliotheque et du codon fausse, pour tester

```python
traducteur = {"AUG": "M", "UUU": "F"}
codon_test = "XYZ"
```

Utilisation de la fonction try et except pour eviter les erreurs, par exemple comme utiliser un codon inconnu dans la bibliotheques 

```python
try: 
    resultat = traducteur[codon_test]
    print(f"Acide aminé : {resultat}")
except KeyError:
    print(f"L'élément '{codon_test}' n'est pas un codon valide dans notre base.")
```
